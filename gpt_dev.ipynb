## ðŸ§  1. What Is ChatGPT?

ChatGPT is a **language model** â€” a neural network trained to predict the next word (or token) in a sequence.
If you give it â€œRoses are red,â€ it learns to continue with â€œviolets are blue.â€

This is called **language modeling.**
Think of it as autocomplete on steroids: it tries to complete your text, one token at a time.

ðŸ‘‰ When you ask ChatGPT to write a poem or answer a question, it doesnâ€™t â€œlook upâ€ answers â€” it *predicts* plausible continuations based on all the text itâ€™s seen.

---

## ðŸ“œ 2. Under the Hood: The Transformer

In 2017, Google released the paper **â€œAttention Is All You Need.â€**
It introduced a neural architecture called the **Transformer**, which powers GPT (Generative Pre-trained Transformer).

Transformers changed deep learning because they replaced recurrence (used in RNNs/LSTMs) with a new idea: **attention.**

> Instead of reading words one by one like a tape recorder, the Transformer looks at the whole sentence and *decides which words to pay attention to* when processing each token.

## âœï¸ 3. Building a Mini-GPT (nanoGPT Style)

To understand how GPT works, Karpathy shows how to train a **tiny Transformer** from scratch â€” on **Shakespeareâ€™s text**.

### ðŸ§© Step 1: Get the Data

* All Shakespeareâ€™s works â†’ one big string of ~1 million characters.
* Vocabulary = unique characters (â‰ˆ 65 total: letters, spaces, punctuation).
* Each character is mapped to a number (a **token**).

### âš™ï¸ Step 2: Tokenization

* Turn text â†’ integers.
* Example:

  ```
  "hi!" â†’ [46, 47, 2]
  ```
* In real GPT models, tokenization happens at a *sub-word* level (like â€œhappiâ€ + â€œnessâ€), using algorithms like **Byte Pair Encoding (BPE).**

### ðŸ§® Step 3: Training Data

We split the data into chunks of length N (â€œblock sizeâ€).
Each chunk teaches the model:

> Given these N characters, predict the next one.

So the model learns: context â†’ next character.

---

## ðŸ§± 4. The Simplest Language Model: Bigram

Start with a simple baseline â€” a **bigram model** â€” which learns only â€œwhat letter follows what.â€

* If it sees â€œth,â€ it learns that â€œeâ€ often follows.
* No deep context, just local patterns.

Loss â‰ˆ how wrong the modelâ€™s predictions are.
Lower loss â†’ better predictions.

---

## ðŸ”„ 5. From Bigram to Transformer

Now we want **tokens to talk to each other** â€” to learn *context.*

Thatâ€™s where **self-attention** comes in.

---

## ðŸŽ¯ 6. Self-Attention: Let Tokens Talk

Imagine every word (or character) as a node that can â€œlookâ€ at other words to decide whatâ€™s relevant.

Each token emits:

* **Query (Q):** â€œWhat am I looking for?â€
* **Key (K):** â€œWhat do I contain?â€
* **Value (V):** â€œWhat information do I send if you look at me?â€

Each tokenâ€™s *attention score* = dot product between its query and othersâ€™ keys.
Then we normalize these scores (softmax) and compute a **weighted average of values**.

ðŸ‘‰ This gives each token a *context-aware* representation â€” it knows which other tokens matter.

---

## ðŸ§© 7. Multi-Headed Attention

One â€œheadâ€ can learn one kind of relationship (like grammar).
Multiple heads learn *different* relationships simultaneously:

* Head 1: who â†’ does what
* Head 2: subject â†” verb
* Head 3: emotion â†” tone

All heads run in parallel, then their outputs are concatenated.

---

## ðŸ§® 8. Feed-Forward + Residuals + Layer Norm

After attention, each token processes its new information with a small **feed-forward network** (an MLP).

Then we add:

* **Residual connections** (skip connections) â€” help gradients flow, avoid vanishing.
* **Layer normalization** â€” stabilizes training.

Each block â‰ˆ â€œcommunicate (attention) â†’ compute (feed-forward).â€
Stack many of them â†’ a **Transformer.**

---

## ðŸ§  9. Scaling Up = GPT

Now scale everything:

* More layers (12 â†’ 96)
* Bigger embeddings (768 â†’ 12 288)
* Larger datasets (1 M tokens â†’ 300 B+)
* Many GPUs

Thatâ€™s **GPT-3 (175 B parameters)**.

## ðŸª„ 10. From Pre-Training to ChatGPT

1. **Pre-training:**
   Learn to predict next token (just like our Shakespeare model) â€” on a large slice of the internet.

2. **Fine-tuning:**
   Teach the model to act like a helpful assistant.
   Humans write good responses â†’ model learns to imitate.

3. **Reinforcement Learning from Human Feedback (RLHF):**
   Humans rank responses â†’ train a *reward model* â†’ optimize GPT to generate high-reward replies.
   This makes ChatGPT polite, safe, and on-topic.

Thatâ€™s why GPT models can *generate* new text, not just analyze it.

````python
import torch                          # PyTorch tensor + autograd library
import torch.nn as nn                 # Neural network layers, modules, etc.
from torch.nn import functional as F  # Functional API (losses, activations, etc.)
torch.manual_seed(1337)               # Make randomness reproducible for this run

class BigramLanguageModel(nn.Module):
    """
    A minimal bigram language model:
    - Learns P(next_token | current_token)
    - Uses an embedding table whose rows are tokens and whose columns are logits over tokens.
    """

    def __init__(self, vocab_size):
        super().__init__()  # Initialize nn.Module internals

        # Embedding table: weight shape == (vocab_size, vocab_size)
        # Think of it as a lookup: given a token id t (row index), return a length-vocab vector of logits.
        # Each row 't' encodes logits for "what comes after token t?"
        self.token_embedding_table = nn.Embedding(vocab_size, vocab_size)

    def forward(self, idx, targets=None):
        """
        idx:      LongTensor of token ids with shape (B, T)
                  B = batch size, T = time steps (sequence length)
        targets:  LongTensor of next-token ids with shape (B, T), optional
        returns:
          logits: FloatTensor of shape (B, T, C) where C = vocab_size
          loss:   Cross-entropy over all (B*T) positions, or None if targets is None
        """

        # 1) Lookup logits for each token in idx:
        #    token_embedding_table does: for every id in idx, return row vector of length vocab_size
        #    idx shape:    (B, T)
        # -> logits shape: (B, T, C)  where C = vocab_size
        logits = self.token_embedding_table(idx)  # (B, T, C)

        if targets is None:
            # If weâ€™re just doing a forward pass for generation, no loss is computed.
            loss = None
        else:
            # Cross-entropy expects 2D logits (N, C) and 1D targets (N)
            # Here N = B*T (flatten all time positions across the batch)

            B, T, C = logits.shape        # Unpack for clarity
            logits = logits.view(B*T, C)  # (B*T, C) -> each row is one time-positionâ€™s logits
            targets = targets.view(B*T)   # (B*T,)   -> each entry is the correct next token id

            # Compute average CE loss over all B*T positions
            loss = F.cross_entropy(logits, targets)

        # Return the unflattened logits if targets was None, or flattened logits if it wasnâ€™t.
        # (Most training loops only consume `loss`; logits shape differences usually donâ€™t matter.)
        return logits, loss

    def generate(self, idx, max_new_tokens):
        """
        Autoregressively sample new tokens.

        idx:            LongTensor with shape (B, T_start)
        max_new_tokens: int, how many steps to extend each sequence

        Returns:
            LongTensor of shape (B, T_start + max_new_tokens)
        """
        # Keep appending one token at a time
        for _ in range(max_new_tokens):
            # Forward pass to get next-token logits for every time step in the current context
            # idx shape:   (B, T_now)
            # logits:      (B, T_now, C)
            # loss:        None (no targets during generation)
            logits, loss = self(idx)

            # Focus only on the logits for the last time step:
            # logits[:, -1, :] picks the final T positionâ€™s distribution
            # shape becomes: (B, C)
            logits = logits[:, -1, :]  # (B, C)

            # Turn logits into probabilities across the vocabulary
            # probs shape: (B, C)
            probs = F.softmax(logits, dim=-1)

            # Sample one token id per batch element from the categorical distribution
            # idx_next shape: (B, 1)
            idx_next = torch.multinomial(probs, num_samples=1)

            # Concatenate along the time dimension to extend the sequence by 1
            # Before: idx shape (B, T_now)
            # After:  idx shape (B, T_now + 1)
            idx = torch.cat((idx, idx_next), dim=1)

        # After the loop, idx has length T_start + max_new_tokens
        return idx

# ----------------------------
# Example usage (assumes these symbols exist in your environment):
# - vocab_size: int
# - xb: (B, T) LongTensor of input token ids
# - yb: (B, T) LongTensor of target token ids
# - decode: function that maps a list of token ids -> string
# ----------------------------

m = BigramLanguageModel(vocab_size)   # Create model; embedding weight shape (vocab_size, vocab_size)

# Forward pass on a batch:
# xb shape: (B, T), yb shape: (B, T)
# logits shape returned here is (B*T, C) because we computed loss and flattened inside forward
logits, loss = m(xb, yb)

print(logits.shape)  # Expect torch.Size([B*T, C])
print(loss)          # Scalar tensor (e.g., tensor(â€¦)) â€” average CE over all positions

# Generation demo:
# Start from a single BOS-like token id 0:
# idx start shape: (1, 1)
generated = m.generate(idx=torch.zeros((1, 1), dtype=torch.long),
                       max_new_tokens=100)  # -> (1, 101)

# Convert token ids to text (user-provided `decode`)
print(decode(generated[0].tolist()))
````

````python
# toy example illustrating how matrix multiplication can be used for a "weighted aggregation"
torch.manual_seed(42)
a = torch.tril(torch.ones(3, 3))
a = a / torch.sum(a, 1, keepdim=True)
b = torch.randint(0,10,(3,2)).float()
c = a @ b
print('a=')
print(a)
print('--')
print('b=')
print(b)
print('--')
print('c=')
print(c)
````

````python
# consider the following toy example:

torch.manual_seed(1337)
B,T,C = 4,8,2 # batch, time, channels
x = torch.randn(B,T,C)
x.shape
````

````python
# We want x[b,t] = mean_{i<=t} x[b,i]
xbow = torch.zeros((B,T,C))
for b in range(B):
    for t in range(T):
        xprev = x[b,:t+1] # (t,C)
        xbow[b,t] = torch.mean(xprev, 0)
````

````python
# version 2: using matrix multiply for a weighted aggregation
wei = torch.tril(torch.ones(T, T))
wei = wei / wei.sum(1, keepdim=True)
xbow2 = wei @ x # (B, T, T) @ (B, T, C) ----> (B, T, C)
torch.allclose(xbow, xbow2)
````

````python
# version 3: use Softmax
tril = torch.tril(torch.ones(T, T))
wei = torch.zeros((T,T))
wei = wei.masked_fill(tril == 0, float('-inf'))
wei = F.softmax(wei, dim=-1)
xbow3 = wei @ x
torch.allclose(xbow, xbow3)
````

````python
# version 4: self-attention!
torch.manual_seed(1337)
B,T,C = 4,8,32 # batch, time, channels
x = torch.randn(B,T,C)

# let's see a single Head perform self-attention
head_size = 16
key = nn.Linear(C, head_size, bias=False)
query = nn.Linear(C, head_size, bias=False)
value = nn.Linear(C, head_size, bias=False)
k = key(x)   # (B, T, 16)
q = query(x) # (B, T, 16)
wei =  q @ k.transpose(-2, -1) # (B, T, 16) @ (B, 16, T) ---> (B, T, T)

tril = torch.tril(torch.ones(T, T))
#wei = torch.zeros((T,T))
wei = wei.masked_fill(tril == 0, float('-inf'))
wei = F.softmax(wei, dim=-1)

v = value(x)
out = wei @ v
#out = wei @ x

out.shape
````

Notes:
- Attention is a **communication mechanism**. Can be seen as nodes in a directed graph looking at each other and aggregating information with a weighted sum from all nodes that point to them, with data-dependent weights.
- There is no notion of space. Attention simply acts over a set of vectors. This is why we need to positionally encode tokens.
- Each example across batch dimension is of course processed completely independently and never "talk" to each other
- In an "encoder" attention block just delete the single line that does masking with `tril`, allowing all tokens to communicate. This block here is called a "decoder" attention block because it has triangular masking, and is usually used in autoregressive settings, like language modeling.
- "self-attention" just means that the keys and values are produced from the same source as queries. In "cross-attention", the queries still get produced from x, but the keys and values come from some other, external source (e.g. an encoder module)
- "Scaled" attention additional divides `wei` by 1/sqrt(head_size). This makes it so when input Q,K are unit variance, wei will be unit variance too and Softmax will stay diffuse and not saturate too much. Illustration below

````python
class LayerNorm1d: # (used to be BatchNorm1d)

  def __init__(self, dim, eps=1e-5, momentum=0.1):
    self.eps = eps
    self.gamma = torch.ones(dim)
    self.beta = torch.zeros(dim)

  def __call__(self, x):
    # calculate the forward pass
    xmean = x.mean(1, keepdim=True) # batch mean
    xvar = x.var(1, keepdim=True) # batch variance
    xhat = (x - xmean) / torch.sqrt(xvar + self.eps) # normalize to unit variance
    self.out = self.gamma * xhat + self.beta
    return self.out

  def parameters(self):
    return [self.gamma, self.beta]

torch.manual_seed(1337)
module = LayerNorm1d(100)
x = torch.randn(32, 100) # batch size 32 of 100-dimensional vectors
x = module(x)
x.shape
````

````python
import torch
import torch.nn as nn
from torch.nn import functional as F

# hyperparameters
batch_size = 16 # how many independent sequences will we process in parallel?
block_size = 32 # what is the maximum context length for predictions?
max_iters = 5000
eval_interval = 100
learning_rate = 1e-3
device = 'cuda' if torch.cuda.is_available() else 'cpu'
eval_iters = 200
n_embd = 64
n_head = 4
n_layer = 4
dropout = 0.0
# ------------

torch.manual_seed(1337)

# wget https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt
with open('input.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# here are all the unique characters that occur in this text
chars = sorted(list(set(text)))
vocab_size = len(chars)
# create a mapping from characters to integers
stoi = { ch:i for i,ch in enumerate(chars) }
itos = { i:ch for i,ch in enumerate(chars) }
encode = lambda s: [stoi[c] for c in s] # encoder: take a string, output a list of integers
decode = lambda l: ''.join([itos[i] for i in l]) # decoder: take a list of integers, output a string

# Train and test splits
data = torch.tensor(encode(text), dtype=torch.long)
n = int(0.9*len(data)) # first 90% will be train, rest val
train_data = data[:n]
val_data = data[n:]

# data loading
def get_batch(split):
    # generate a small batch of data of inputs x and targets y
    data = train_data if split == 'train' else val_data
    ix = torch.randint(len(data) - block_size, (batch_size,))
    x = torch.stack([data[i:i+block_size] for i in ix])
    y = torch.stack([data[i+1:i+block_size+1] for i in ix])
    x, y = x.to(device), y.to(device)
    return x, y

@torch.no_grad()
def estimate_loss():
    out = {}
    model.eval()
    for split in ['train', 'val']:
        losses = torch.zeros(eval_iters)
        for k in range(eval_iters):
            X, Y = get_batch(split)
            logits, loss = model(X, Y)
            losses[k] = loss.item()
        out[split] = losses.mean()
    model.train()
    return out

class Head(nn.Module):
    """ one head of self-attention """

    def __init__(self, head_size):
        super().__init__()
        self.key = nn.Linear(n_embd, head_size, bias=False)
        self.query = nn.Linear(n_embd, head_size, bias=False)
        self.value = nn.Linear(n_embd, head_size, bias=False)
        self.register_buffer('tril', torch.tril(torch.ones(block_size, block_size)))

        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        B,T,C = x.shape
        k = self.key(x)   # (B,T,C)
        q = self.query(x) # (B,T,C)
        # compute attention scores ("affinities")
        wei = q @ k.transpose(-2,-1) * C**-0.5 # (B, T, C) @ (B, C, T) -> (B, T, T)
        wei = wei.masked_fill(self.tril[:T, :T] == 0, float('-inf')) # (B, T, T)
        wei = F.softmax(wei, dim=-1) # (B, T, T)
        wei = self.dropout(wei)
        # perform the weighted aggregation of the values
        v = self.value(x) # (B,T,C)
        out = wei @ v # (B, T, T) @ (B, T, C) -> (B, T, C)
        return out



````

# One Attention Head (Causal Self-Attention) â€” Intuition + Tiny Example (No LaTeX)

**What does a â€œheadâ€ do?**
For each position *t* in a sequence, the head asks: â€œWhich earlier tokens matter to me, and by how much?â€
It projects inputs into three spacesâ€”Query (Q), Key (K), and Value (V)â€”scores relevance with dot products, **masks the future**, softmaxes into probabilities, then blends the Values accordingly.

---

## 1) The three learned lenses

From input embeddings `x` with shape `B Ã— T Ã— n_embd`, we learn three linear projections to `head_size = C`:

* `Q = x Â· W_Q`
* `K = x Â· W_K`
* `V = x Â· W_V`

**Intuition**

* **Query (Q)** = what Iâ€™m looking for
* **Key (K)** = what I offer
* **Value (V)** = the content Iâ€™ll share if you attend to me

---

## 2) Scoring, masking, and mixing

1. **Affinity scores** (how much token *i* looks at token *j*):

```
scores = (Q @ K^T) / sqrt(C)    # shape: B Ã— T Ã— T
```

2. **Causal mask** (no peeking ahead): set the upper triangle to `-inf` so softmax ignores the future.

3. **Softmax** row-wise:

```
weights = softmax(scores, dim=-1)
```

4. **Weighted sum of values**:

```
out = weights @ V                # shape: B Ã— T Ã— C
```

# Multi-Head Attention â€” Many â€œmini-attentionsâ€ in parallel (simple, visual)

**Whatâ€™s the idea?**
Instead of one attention head, we run **several heads in parallel**.
Each head learns to focus on *different* patterns (syntax, long-range links, punctuation, etc.).
Then we **concatenate** their outputs and use a **linear projection** to mix them back into the model dimension.

---

## The class at a glance

```text
class MultiHeadAttention(nn.Module):
  heads   = [Head(head_size) for _ in range(num_heads)]   # parallel heads
  proj    = nn.Linear(n_embd, n_embd)                     # mix heads together
  dropout = nn.Dropout(dropout)

forward(x):
  out = concat([h(x) for h in heads], dim=-1)  # join along last dim
  out = dropout(proj(out))                     # mix + regularize
  return out
```

* `Head(head_size)` is the same causal self-attention unit you just learned.
* `torch.cat(..., dim=-1)` glues head outputs side-by-side across the **channel** dimension.
* `proj` brings the concatenated vector (size `num_heads * head_size`) back to `n_embd` and lets heads â€œtalkâ€ to each other.
* `dropout` is regularization on the mixed output.

**Shape rule of thumb**

```
Input x:           (B, T, n_embd)
Each head output:  (B, T, head_size)
Concat all heads:  (B, T, num_heads * head_size)  â† should equal n_embd
Projection:        Linear(n_embd â†’ n_embd)
Final:             (B, T, n_embd)
```

---

Letâ€™s make the shapes small and clear:

[

    [0.80, 0.20] ]  # t3
]
```

### 1) Concatenate along the last dimension (`dim=-1`)

We glue the per-token vectors from head 1 and head 2 side-by-side:

```
concat = torch.cat([head1, head2], dim=-1)  # shape (B,T,4)

concat =
[
  [ [1.00, 0.00, 0.10, 0.90],   # t1: [h1..., h2...]
    [1.00, 0.73, 0.40, 0.60],   # t2
    [0.58, 0.84, 0.80, 0.20] ]  # t3
]
```

Now each token has a 4-D vector (2 dims per head Ã— 2 heads).

### 2) Mix heads with a projection (`nn.Linear(4, 4)`)

To keep it simple, imagine `proj` is currently an identity-like mapping (so we can see shapes clearly). In a real model, `proj` is learned and can **blend** the head features:

```
out = proj(concat)   # still (B,T,4) but now â€œheads can talk to each otherâ€
out = dropout(out)   # regularization during training
```

**Intuition:**

* Head 1 might specialize in â€œwho refers to whoâ€ (coreference).
* Head 2 might specialize in â€œnearby collocationsâ€ (local patterns).
* `proj` combines these perspectives into one `n_embd`-dim vector per token.

---

## Why multi-heads help

* **Different views:** Each head has its own Q/K/V weights â†’ different attention patterns.
* **Richer features:** Concatenation preserves each headâ€™s unique info before mixing.
* **Controlled capacity:** `num_heads * head_size` is kept equal to `n_embd`, so the model width stays constant.

```

# Whatâ€™s a **bigram language model**?

A **bigram LM** predicts the next token using **only the previous token**.
Formally: â€œthe probability of word *wáµ¢* depends only on *wáµ¢â‚‹â‚*.â€

### Tiny example

Training text:
`<BOS> i like pizza <EOS>`
`<BOS> i like pasta <EOS>`
`<BOS> you like pizza <EOS>`

Count adjacent pairs (bigrams):

```
<BOS>â†’i : 3
iâ†’like  : 2
youâ†’like: 1
likeâ†’pizza: 2
likeâ†’pasta: 1
pizzaâ†’<EOS>: 2
pastaâ†’<EOS>: 1
```

Turn counts into probabilities by normalizing **outgoing** edges:

```
P(i | <BOS>)   = 3 / 3 = 1.0
P(like | i)    = 2 / 2 = 1.0
P(like | you)  = 1 / 1 = 1.0
P(pizza | like)= 2 / 3 â‰ˆ 0.67
P(pasta | like)= 1 / 3 â‰ˆ 0.33
P(<EOS>| pizza)= 2 / 2 = 1.0
P(<EOS>| pasta)= 1 / 1 = 1.0
```

To score a sentence like `<BOS> i like pizza <EOS>`, multiply the bigram probabilities along the path.

> In practice you add **smoothing** (e.g., add-k, Kneser-Ney) so unseen pairs donâ€™t get zero probability.

### Pros / cons

* âœ… Simple, fast, interpretable; great teaching baseline.
* âŒ Short memory (just one token!), canâ€™t capture long-range structure.

# Other types of language models

## 1) N-gram (count-based) family

* **Unigram**: next token ignores history entirely (just frequencies).
* **Bigram / Trigram / 4-gram+**: depend on last 1/2/3â€¦ tokens.
* **Smoothing & backoff**: Laplace/add-k, **Kneser-Ney**, Katz, interpolated backoff.
* **Caches / interpolations**: mix a global LM with a short-term cache of recent words.

## 2) Early neural LMs

* **Feed-forward NNLM** (Bengio et al., 2003): fixed window of recent tokens â†’ MLP â†’ next-token.
* **Class/factorized softmax**: speedups for huge vocabularies.

## 3) Recurrent models

* **RNN LM**: maintains a hidden state over the whole prefix.
* **LSTM / GRU**: better at long-range dependencies than vanilla RNNs.
* Variants: **char-level** (predict characters), **word/subword-level** (BPE, WordPiece).

## 4) Convolutional LMs

* **CNN LMs** (e.g., gated CNN): use dilated convolutions to capture longer contexts with parallelism.

## 5) Attention-based (Transformers)

* **Causal (autoregressive)**: predict token *t* from tokens `< t`. Examples: GPT-style models; used for generation.
* **Masked (bidirectional)**: fill in masked tokens using both left and right context. Example: BERT (great for understanding/encoding, not pure generation).
* **Seq2Seq (encoderâ€“decoder)**: encode a source sequence, decode a target (e.g., translation, summarization). Examples: T5, BART.
* **Mixture-of-Experts (MoE)**: route tokens to specialized expert subnets for scale.

## 6) Retrieval-augmented & hybrids

* **RAG / RETRO / kNN-LM**: consult an external datastore during prediction to reduce hallucinations and improve factuality.

## 7) Diffusion / energy-based / other researchy LMs

* Alternative training objectives (less common in everyday NLP, active research).

# How they differ (at a glance)

```
Memory     | Unigram  Bigram  Trigram  N-gram+  RNN/LSTM/GRU  Transformer
-----------|--------------------------------------------------------------
Context    | none     1 token 2 tokens few tok  long (hidden) long (self-attn)
Parallel   | high     high    high     high     low            high
Generalize | low      low     medium   medium   high           very high
```

# Practical notes

* **Tokenization matters**: modern LMs use subwords (BPE/WordPiece) to balance vocab size and OOV handling.
* **Evaluation**: perplexity (lower is better) compares model fit on held-out text.
* **When to use what**: n-grams for tiny data or pedagogy; RNNs if you must be lightweight; Transformers for state-of-the-art performance.

````python
class FeedFoward(nn.Module):
    """ a simple linear layer followed by a non-linearity """

    def __init__(self, n_embd):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(n_embd, 4 * n_embd),
            nn.ReLU(),
            nn.Linear(4 * n_embd, n_embd),
            nn.Dropout(dropout),
        )

    def forward(self, x):
        return self.net(x)

class Block(nn.Module):
    """ Transformer block: communication followed by computation """

    def __init__(self, n_embd, n_head):
        # n_embd: embedding dimension, n_head: the number of heads we'd like
        super().__init__()
        head_size = n_embd // n_head
        self.sa = MultiHeadAttention(n_head, head_size)
        self.ffwd = FeedFoward(n_embd)
        self.ln1 = nn.LayerNorm(n_embd)
        self.ln2 = nn.LayerNorm(n_embd)

    def forward(self, x):
        x = x + self.sa(self.ln1(x))
        x = x + self.ffwd(self.ln2(x))
        return x

# super simple bigram model
class BigramLanguageModel(nn.Module):

    def __init__(self):
        super().__init__()
        # each token directly reads off the logits for the next token from a lookup table
        self.token_embedding_table = nn.Embedding(vocab_size, n_embd)
        self.position_embedding_table = nn.Embedding(block_size, n_embd)
        self.blocks = nn.Sequential(*[Block(n_embd, n_head=n_head) for _ in range(n_layer)])
        self.ln_f = nn.LayerNorm(n_embd) # final layer norm
        self.lm_head = nn.Linear(n_embd, vocab_size)

    def forward(self, idx, targets=None):
        B, T = idx.shape

        # idx and targets are both (B,T) tensor of integers
        tok_emb = self.token_embedding_table(idx) # (B,T,C)
        pos_emb = self.position_embedding_table(torch.arange(T, device=device)) # (T,C)
        x = tok_emb + pos_emb # (B,T,C)
        x = self.blocks(x) # (B,T,C)
        x = self.ln_f(x) # (B,T,C)
        logits = self.lm_head(x) # (B,T,vocab_size)

        if targets is None:
            loss = None
        else:
            B, T, C = logits.shape
            logits = logits.view(B*T, C)
            targets = targets.view(B*T)
            loss = F.cross_entropy(logits, targets)

        return logits, loss

    def generate(self, idx, max_new_tokens):
        # idx is (B, T) array of indices in the current context
        for _ in range(max_new_tokens):
            # crop idx to the last block_size tokens
            idx_cond = idx[:, -block_size:]
            # get the predictions
            logits, loss = self(idx_cond)
            # focus only on the last time step
            logits = logits[:, -1, :] # becomes (B, C)
            # apply softmax to get probabilities
            probs = F.softmax(logits, dim=-1) # (B, C)
            # sample from the distribution
            idx_next = torch.multinomial(probs, num_samples=1) # (B, 1)
            # append sampled index to the running sequence
            idx = torch.cat((idx, idx_next), dim=1) # (B, T+1)
        return idx

model = BigramLanguageModel()
m = model.to(device)
# print the number of parameters in the model
print(sum(p.numel() for p in m.parameters())/1e6, 'M parameters')

# create a PyTorch optimizer
optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate)

for iter in range(max_iters):

    # every once in a while evaluate the loss on train and val sets
    if iter % eval_interval == 0 or iter == max_iters - 1:
        losses = estimate_loss()
        print(f"step {iter}: train loss {losses['train']:.4f}, val loss {losses['val']:.4f}")

    # sample a batch of data
    xb, yb = get_batch('train')

    # evaluate the loss
    logits, loss = model(xb, yb)
    optimizer.zero_grad(set_to_none=True)
    loss.backward()
    optimizer.step()

# generate from the model
context = torch.zeros((1, 1), dtype=torch.long, device=device)
print(decode(m.generate(context, max_new_tokens=2000)[0].tolist()))
````
